```python
from pylab import *
from matplotlib import rc
import matplotlib.gridspec as gridspec
import numpy as np
```

```python
%matplotlib notebook
clf()
xoff=0.22
yoff=0.2
width=0.8-xoff
height=0.95-yoff


fig_width_pt = 246.0
inches_per_pt = 1.0/72.27
golden_mean = (sqrt(5)-1.0)/2.0
fig_width = fig_width_pt*inches_per_pt
fig_height = fig_width_pt*inches_per_pt*golden_mean
fig_size =  [2.5*fig_width,2.5*fig_height]
params = {'backend': 'pdf',
          'axes.labelsize': 9,
          'xtick.labelsize': 8,
          'ytick.labelsize': 8,
          'text.usetex': False,
          'figure.figsize': fig_size,
          'figure.subplot.left': xoff,
          'figure.subplot.bottom': yoff,
          'figure.subplot.right': xoff+width,
          'figure.subplot.top': yoff+height,
          'mathtext.fontset': 'stixsans'}

rcParams.update(params)
rc('font',**{'family':'sans-serif','sans-serif':['Tex Gyre Heros']})
        
figure(1)

gs1=gridspec.GridSpec(2,2)
gs1.update(hspace=0.0,wspace=0.2)

abs1=np.loadtxt('asys1_abs.txt',delimiter='\t')
abs2=np.loadtxt('asys2_abs.txt',delimiter='\t')
vg1=np.loadtxt('asys1_vg.txt',delimiter='\t')
vg2=np.loadtxt('asys2_vg.txt',delimiter='\t')

subplot(gs1[0,0])

plot([e[0] for e in abs1],[e[1] for e in abs1],color='b',lw=1)
ll,=plot([496.0,496.0],[0,1.05],color='b',lw=1)
ll.set_dashes([2,2])

setp(gca(),xlim=[400,600],ylim=[0,1.05],xticklabels=[])
title('Ag(30 nm)-TiO$_2$(50 nm)-Ag(100 nm)',fontsize=9,color='b')
ylabel('Absorption $A$')
text(407,0.95,'A',fontsize=10,fontweight='bold')

subplot(gs1[1,0])
plot([e[0] for e in vg1],[e[1] for e in vg1],color='b',lw=1)
ll,=plot([496.0,496.0],[0.05,20.0],color='b',lw=1)
ll.set_dashes([2,2])

gca().set_yscale('log')
setp(gca(),xlim=[400,600],ylim=[0.05,20.0])
ylabel('$v_g\, / c$')
xlabel('Wavelength [nm]')
text(407,10.5,'B',fontsize=10,fontweight='bold')

subplot(gs1[0,1])

plot([e[0] for e in abs2],[e[1] for e in abs2],color='r',lw=1)
ll,=plot([474.0,474.0],[0,1.05],color='r',lw=1)
ll.set_dashes([2,2])

setp(gca(),xlim=[400,600],ylim=[0,1.05],xticklabels=[])
title('Ge(15 nm)-Ag(30 nm)-TiO$_2$(50 nm)-Ag(100 nm)',fontsize=9,color='r')
text(407,0.95,'C',fontsize=10,fontweight='bold')

subplot(gs1[1,1])
plot([e[0] for e in vg2],[e[1] for e in vg2],color='r',lw=1)
ll,=plot([474.0,474.0],[0.05,20.0],color='r',lw=1)
ll.set_dashes([2,2])

gca().set_yscale('log')
setp(gca(),xlim=[400,600],ylim=[0.05,20.0])
xlabel('Wavelength [nm]')
text(407,10.5,'D',fontsize=10,fontweight='bold')

show()
savefig('vgplots.pdf',bbox_inches='tight')
```

```python
%matplotlib notebook
clf()
xoff=0.22
yoff=0.2
width=0.8-xoff
height=0.95-yoff


fig_width_pt = 246.0
inches_per_pt = 1.0/72.27
golden_mean = (sqrt(5)-1.0)/2.0
fig_width = fig_width_pt*inches_per_pt
fig_height = fig_width_pt*inches_per_pt*golden_mean
fig_size =  [1*fig_width,2.5*fig_height]
params = {'backend': 'pdf',
          'axes.labelsize': 9,
          'xtick.labelsize': 8,
          'ytick.labelsize': 8,
          'text.usetex': False,
          'figure.figsize': fig_size,
          'figure.subplot.left': xoff,
          'figure.subplot.bottom': yoff,
          'figure.subplot.right': xoff+width,
          'figure.subplot.top': yoff+height,
          'mathtext.fontset': 'stixsans'}

rcParams.update(params)
rc('font',**{'family':'sans-serif','sans-serif':['Tex Gyre Heros']})
        
figure(1)

gs1=gridspec.GridSpec(3,1)
gs1.update(hspace=0.0,wspace=0.0)

osc1i=np.loadtxt('osc1_intensity.txt',delimiter='\t')
osc2i=np.loadtxt('osc2_intensity.txt',delimiter='\t')
osc1p=np.loadtxt('osc1_phase.txt',delimiter='\t')
osc2p=np.loadtxt('osc2_phase.txt',delimiter='\t')
osc1r=np.loadtxt('osc1_refl.txt',delimiter='\t')
osc1cr=np.loadtxt('osc1_coupled_refl.txt',delimiter='\t')

subplot(gs1[0,0])

plot([e[0] for e in osc1i],[e[1] for e in osc1i],color='b',lw=1)
plot([e[0] for e in osc2i],[e[1] for e in osc2i],color='r',lw=1)
l1,=plot([2.68,2.68],[1,5e3],color='k',lw=0.75)
l1.set_dashes([2,2])
l1,=plot([3.468,3.468],[1,5e3],color='k',lw=0.75)
l1.set_dashes([2,2])
gca().set_yscale('log')

setp(gca(),xlim=[2.0,4.0],ylim=[1,5e3],xticklabels=[])
ylabel('Intensity $|A(\omega)|^2$')
gca().tick_params(axis='x',which='both',direction="in")
gca().yaxis.set_label_coords(-0.19,0.5)


subplot(gs1[1,0])

plot([e[0] for e in osc1p],[e[1] for e in osc1p],color='b',lw=1)
plot([e[0] for e in osc2p],[e[1] for e in osc2p],color='r',lw=1)
plot([2,4],[0,0],color='k',lw=0.75)
l1,=plot([2,4],[3.1415/2,3.1415/2],color='k',lw=0.75)
l1.set_dashes([2,2])
l2,=plot([2,4],[-3.1415/2,-3.1415/2],color='k',lw=0.75)
l2.set_dashes([2,2])

setp(gca(),xlim=[2.0,4.0],ylim=[-2,2],xticklabels=[],yticks=arange(-3.1415/2,3.1415,3.1415/2),yticklabels=[r'$-\pi/2$',r'0',r'$\pi/2$'])
ylabel('Phase $\Phi(\omega)$ [rad]')
gca().tick_params(axis='x',which='both',direction="in")
gca().yaxis.set_label_coords(-0.19,0.5)

subplot(gs1[2,0])

plot([e[0] for e in osc1r],[e[1] for e in osc1r],color='b',lw=1)
plot([e[0] for e in osc1cr],[e[1] for e in osc1cr],color='g',lw=1)
setp(gca(),xlim=[2.0,4.0],ylim=[0,0.8],xticks=arange(2,4.5,0.5))
ylabel('Reflectance')
xlabel('Angular freq. $\omega$ [$10^{15}$ Hz]')
gca().yaxis.set_label_coords(-0.19,0.5)

show()
savefig('oscillators.pdf',bbox_inches='tight')
```
